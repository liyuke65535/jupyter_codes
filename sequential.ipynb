```python
from tensorflow import keras
import numpy as np

# sequential线性模型。units: 正整数，输出空间维度。
# input_shape: 例：[1,2,3] shape 为 (1,);   [[1],[2],[3]] 这个张量的shape为(3,1)

model=keras.Sequential([keras.layers.Dense(units=1,input_shape=[1])]) 
model.compile(optimizer='sgd',loss='mean_squared_error')

xs=np.array([-1,-2,12,3.4,1,-8,2,0,3,0,4,30],dtype=float) # 输入
ys=np.array([-3,-5,23,5.8,1,-19,3,0,5,0,7,59],dtype=float) # 输出

model.fit(xs,ys,epochs=200) #训练次数epochs
```

```python
model.predict([19.0]) # 预测
```
