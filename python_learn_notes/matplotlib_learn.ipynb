# 一.Pyplot
- Pyplot 是 Matplotlib 的子库，提供了和 MATLAB 类似的绘图 API。
- Pyplot 是常用的绘图模块，能很方便让用户绘制 2D 图表。
- Pyplot 包含一系列绘图函数的相关函数，每个函数会对当前的图像进行一些修改，例如：给图像加上标记，生新的图像，在图像中产生新的绘图区域等等。

## 1.plot

**参数说明：**
- x, y：点或线的节点，x 为 x 轴数据，y 为 y 轴数据，数据可以列表或数组。
- fmt：可选，定义基本格式（如颜色、标记和线条样式）。
- **kwargs：可选，用在二维平面图上，设置指定属性，如标签，线的宽度等。

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 100])

plt.plot(xpoints, ypoints)
plt.show()
```

**fmt参数**
- fmt='[marker][line][color]'
- marker:点标记样式
- line:线样式
- color:线颜色

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([2,0])
y = np.array([0,100])
x1 = np.array([1,1.5,2,2.5])
y1 = np.array([0,10,50,80])
plt.plot(x, y)        # 创建 y 中数据与 x 中对应值的二维线图，使用默认样式
plt.plot(x, y, 'bo')  # 创建 y 中数据与 x 中对应值的二维线图，使用蓝色实心圈(blue o)绘制
plt.plot(y)           # x 的值为 0..N-1
plt.plot(y, 'r+')     # 使用红色 + 号
plt.plot(x1,y1,'>:y') # >右三角 :虚线 y黄色
```

## 2.grid网格线

```python
import numpy as np
import matplotlib.pyplot as plt

y = np.arange(0,25.1,0.1) ** 0.5
plt.plot(y)
plt.grid(color='r', linestyle=':', linewidth=0.5) # 网格线
plt.grid(axis='x') # 只有横线
plt.show()
```

## 3.scatter散点图

```python
import matplotlib.pyplot as plt
import numpy as np
n = 200 # 用于生成十个点
x = np.random.rand(n)
y = np.random.rand(n) * 25
labels = np.random.randint(3,size=n) # 0，1，2组成的随机数串，作为标签，让颜色分类
# print(x,y,labels)

plt.scatter(x, y, c=labels) # 颜色按labels分类
plt.show()
```

## 4.绘制多图
**subplot(nrows, ncols, index, \*\*kwargs)**
- numRows ＝ 1，numCols ＝ 2，就是将图表绘制成 1x2 的图片区域
- index为序号

```python
import matplotlib.pyplot as plt
import numpy as np
#plot 1:
xpoints = np.array([0, 6])
ypoints = np.array([0, 100])

plt.subplot(1, 2, 1) # 绘制成1行2列，这是序号为1的图
plt.plot(xpoints,ypoints)
plt.title("plot 1")

#plot 2:
x = np.array([1, 2, 3, 4])
y = np.array([1, 4, 9, 16])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("plot 2")

plt.suptitle("RUNOOB subplot Test")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# 创建一些测试数据 -- 图1
x = np.linspace(0, 2*np.pi, 200)
y = np.sin(x**2)

# 创建一个画像和子图 -- 图2
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Simple plot')

# 创建两个子图 -- 图3
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(x, y)
ax1.set_title('Sharing Y axis')
ax2.scatter(x, y)

# 创建四个子图 -- 图4
fig, axs = plt.subplots(2, 2, subplot_kw=dict(projection="polar"))
axs[0, 0].plot(x, y)
axs[1, 1].scatter(x, y)

# 共享 x 轴和 y 轴
plt.subplots(2, 2, sharex='all', sharey='all')
'''
# 共享 x 轴
plt.subplots(2, 2, sharex='col')

# 共享 y 轴
plt.subplots(2, 2, sharey='row')

# 这个也是共享 x 轴和 y 轴
plt.subplots(2, 2, sharex=True, sharey=True)

# 创建10 张图，已经存在的则删除
fig, ax = plt.subplots(num=10, clear=True)
'''
plt.show()
```

## 5.bar柱状图

- x：浮点型数组，柱形图的 x 轴数据。
- height：浮点型数组，柱形图的高度。
- width：浮点型数组，柱形图的宽度。
- bottom：浮点型数组，底座的 y 坐标，默认 0。
- align：柱形图与 x 坐标的对齐方式，'center' 以 x 位置为中心，这是默认值。 'edge'：将柱形图的左边缘与 x 位置对齐。要对齐右边缘的条形，可以传递负数的宽度值及 align='edge'。

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["Runoob-1", "Runoob-2", "Runoob-3", "C-RUNOOB"])
y = np.array([12, 22, 6, 18])

# 纵向柱状图只可以设置width，横向只可以设置height
plt.bar(x,y,width=0.1,color = ["#4CAF50","red","hotpink","#556B2F"])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["Runoob-1", "Runoob-2", "Runoob-3", "C-RUNOOB"])
y = np.array([12, 22, 6, 18])

plt.barh(x,y,height=[0.1,0.2,0.4,0.8]) # 垂直方向的bar
plt.show()
```
